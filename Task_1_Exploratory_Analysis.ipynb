  ## Унификация едениц измерения

  Данные представлены в разных еденицах измерения.

  Параметры, в названия которых встроены еденицы измерения:

```python
import re

matched_columns = []
unmatched_columns = []

pattern = r"\((g|mg|µg)\)"

for column in df.columns:
    match = re.search(pattern, column)
    if match:
        matched_columns.append(column)
    else:
        unmatched_columns.append(column)

print("\nMatched Columns:")
print(matched_columns)
print("\nUnmatched Columns:")
print(unmatched_columns)
```

Названия параметров `Copper_mg)`, `Panto_Acid_mg)` и `Vit_D_µg` искажены. Так же в некоторых колонках есть лишние пробелы. Исправим это.

```python
df.rename(columns={'Copper_mg)': 'Copper_(mg)',
                   'Panto_Acid_mg)': 'Panto_Acid_(mg)',
                   'Vit_D_µg': 'Vit_D_(µg)'}, inplace=True)

df.rename(columns={'Choline_Tot_ (mg)': 'Choline_Tot_(mg)',
                   'Lut+Zea_ (µg)': 'Lut+Zea_(µg)'}, inplace=True)

column_units = {}

pattern = r"\((g|mg|µg)\)"

for column in df.columns:
    match = re.search(pattern, column)
    if match:
        column_name = re.sub(pattern, '', column)

        measuring_unit = match.group(1)

        column_units[column_name.strip('_')] = measuring_unit

unit_df = pd.DataFrame(list(column_units.items()),
                       columns=['Параметр', 'Еденица измерения'])

print(unit_df)
```

### Переведём всё в миллиграммы:

```python
unit_conversion = {'g': 1000, 'mg': 1, 'µg': 0.001}

for index, row in unit_df.iterrows():
    column_name = row['Параметр']
    measuring_unit = row['Еденица измерения']
    original_column_name = column_name + '_(' + measuring_unit + ')'

    conversion_factor = unit_conversion[measuring_unit]
    df[original_column_name] *= conversion_factor

    df.rename(columns={original_column_name: column_name}, inplace=True)

df
```

### Параметр 'Sugar_Tot'

  Суммарное количества сахара явно зависит от количества углеводов.
  Поэтому можно более аккуратно заполнить недостающие значения, чем используя средее всей выборки.

```python
sns.scatterplot(x='Carbohydrt', y='Sugar_Tot', data=df, alpha=0.5)
plt.title('Зависимость количества углеводов от содержания сахара')
plt.xlabel('Углеводы')
plt.ylabel('Сахар')
plt.show()
```

Распределим значения `Carbohydrt` по "корзинам" и для каждой "корзины" вычислим среднее значение `Sugar_Tot`. Недостающие значения `Sugar_Tot` заполним на основе соответствующего среднего значения `Carbohydrt`.

```python
num_bins = 10

df['Carbohydrt_bins'] = pd.cut(df['Carbohydrt'], bins=num_bins, labels=False)

mean_sugar_by_carbohydrt = df.groupby('Carbohydrt_bins')['Sugar_Tot'].mean()

def fill_missing_sugar(row):
    if pd.isnull(row['Sugar_Tot']):
        carbohydrt_bin = row['Carbohydrt_bins']
        return mean_sugar_by_carbohydrt[carbohydrt_bin]
    return row['Sugar_Tot']

df['Sugar_Tot'] = df.apply(fill_missing_sugar, axis=1)

if df['Sugar_Tot'].isnull().sum() == 0: print("Success")
```

### Параметры 'Copper', 'Manganese' и 'Selenium'

Данные параметры являются неорганическими.

  Так же мы знаем параметр 'Ash', отвечающий за суммарное количество неорганических соединений в продукте. Золой он называется из-за способа его определения. Образцы сырья помещаются в печь и сжигаются до постоянной массы. В результате остаются только элементы, которые не сгорают, а именно минералы.

  Посмотрим на зависимость между суммой неорганических элементов и параметром 'Ash'

```python
inorganic = ['Calcium', 'Iron', 'Magnesium', 'Phosphorus', 'Potassium', 'Sodium',
             'Zinc', 'Copper', 'Manganese', 'Selenium']

df['Sum_Inorganic'] = df[inorganic].sum(axis=1)

plt.figure(figsize=(10, 6))
sns.lineplot(x='Ash', y='Sum_Inorganic', data=df, marker='o', color='b', linewidth=2)
plt.xlabel('Зола')
plt.ylabel('Суммарное количество неорганических соединений')
plt.show()
```

  Видна явная линейная зависимость, ввиду чего недостающее значение меди, магнезии или селена можно вычислить как разность между золой и суммой остальных неорганических элементов.

  Сумма берётся с коэффициентом 1, $\frac{1}{2}$ или $\frac{1}{3}$ в зависимости от числа недостающих элементов на текущей строке.

```python
missing_values = df['Ash'] - df['Sum_Inorganic']

coefficient = (df['Copper'].isnull().astype(int) +
               df['Manganese'].isnull().astype(int) +
               df['Selenium'].isnull().astype(int))
coefficient = coefficient.apply(lambda x: 0.3 if x == 3 else (0.5 if x == 2 else 1))

df['Copper'] = df['Copper'].fillna(missing_values * coefficient)
df['Manganese'] = df['Manganese'].fillna(missing_values * coefficient)
df['Selenium'] = df['Selenium'].fillna(missing_values * coefficient)

print(df[['Copper', 'Manganese', 'Selenium']].isnull().sum())
```

```python
fatty_acid_means = df.groupby('Protein Category')[['FA_Sat', 'FA_Mono', 'FA_Poly']].mean()

protein_categories = ['Low', 'Medium', 'High']
fatty_acids = ['FA_Sat', 'FA_Mono', 'FA_Poly']

positions = np.arange(len(protein_categories))
bar_width = 0.2

for i, fatty_acid in enumerate(fatty_acids):
    plt.bar(positions + i * bar_width, fatty_acid_means[fatty_acid], bar_width, label=fatty_acid)

plt.xticks(positions + bar_width, protein_categories)
plt.legend(title='Жирные кислоты')
plt.tight_layout()
plt.show()
```

В продуктах с высоким содержанием белка больше всего жирных кислот.
В продуктах с низким содержанием белка меньше всего полиненасыщенных жирных кислот. Параметры `FA_Sat`, `FA_Mono`, и `FA_Poly` важны, однако для уменьшения вычислительной сложности можно использовать суммарное число жирных кислот:

```python
df['Fatty_Acids'] = df['FA_Sat'] + df['FA_Mono'] + df['FA_Poly']

sns.barplot(x='Protein Category', y='Fatty_Acids', data=df, estimator=np.median)

df.drop(columns=['FA_Sat', 'FA_Mono', 'FA_Poly'], inplace=True)
```

### Различные витамины

В датасете присутствуют различные витамины: C, E, D, K, B, A. Посмотрим, имеют ли они хоть какую-то связь с количеством белка:

```python
vitamin_columns = ['Vit_C', 'Vit_E', 'Vit_D', 'Vit_K', 'Vit_B', 'Vit_A']

# Calculate the correlation matrix
correlation_matrix = df[vitamin_columns + ['Protein']].corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Корреляция между белком и витаминами')
plt.show()
```
